# 1.Import the libraries

```python
import cv2
import matplotlib.pyplot as plt
```

# 2.Load the image

```python
image = cv2.imread("C:\\Users\\hp\\OneDrive\\Documents\\LGM_Tasks\\Fish.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# 3.Convert the original image to grayscale

```python
imag_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(imag_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

# 4.Invert the grayscale image

```python
imag_invert = cv2.bitwise_not(imag_gray)
plt.figure(figsize=(7,7))
plt.imshow(imag_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

# 5.Blur the image by using the Gaussian Function in OpenCV

```python
imag_blur = cv2.GaussianBlur(imag_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(7,7))
plt.imshow(imag_blur,cmap="gray")
plt.axis("off")
plt.title("Blur Image")
plt.show()
```

# 6.Invert the blurred image

```python
final_imag = cv2.divide(imag_gray, 255 - imag_blur, scale=255)
plt.figure(figsize=(7,7))
plt.imshow(final_imag,cmap="gray")
plt.axis("off")
plt.title("Pencil Sketch Image")
plt.show()
```
